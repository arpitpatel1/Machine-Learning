Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(xtrain,ytrain, test_size = 0.20, random_state =5)
```

```python
x_test
```

```python
y_test
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

MODEL EVALUATION

1) Decision Tree

```python
from sklearn import tree 
mod1 = tree.DecisionTreeClassifier(max_depth=16)
mod1.fit(x_train, y_train)
mod1.predict(x_test)
```

```python
mod1.score(x_test,y_test)*100
```

2) Random Forest(Max Depth=16)

```python
 from sklearn.ensemble import RandomForestClassifier
 mod2 = RandomForestClassifier(n_estimators=99, max_depth=16, random_state=42)
 mod2.fit(x_train, y_train)
 mod2.predict(x_test)
```

```python
mod2.score(x_test,y_test)*100
```

3) KNN

```python
from sklearn.neighbors import KNeighborsClassifier
mod3 = KNeighborsClassifier(n_neighbors=3)
mod3.fit(x_train, y_train)
```

```python
y_pred=mod3.predict(x_test)
y_pred
```

```python
mod3.score(x_test,y_test)*100
```

4) Random forest(Max depth=32)

```python
 mod4 = RandomForestClassifier(n_estimators=99, max_depth=32, random_state=42)
 mod4.fit(x_train, y_train)
 mod4.predict(x_test)
```

```python
mod4.score(x_test,y_test)*100
```

**CONCLUSION**

In summary, out of the 3 models we tested, we find the most accurate results with our 2nd model i.e Decision tree with a maxdepth 16.

We can also improve on our existing model by implementing methods of data engineering and feature selection during the pre-processing stage. Using neural networks should improve the accuracy and precision of the predictions.
