```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime
```

```python
df = pd.read_csv("heart.csv")
```

```python
df.sample(3)
```

```python
x = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
x.isnull().sum()
```

```python
x.info()
```

```python
x.shape
```

```python
sns.pairplot(x)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(x.corr(),annot=True,fmt='.1f')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.30, random_state =5)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
dtree = DecisionTreeClassifier(random_state=50)
dtree.fit(x_train, y_train)
```

```python
dt = dtree.predict(x_test) 
accuracy_score(y_test,dt)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=2, random_state=50)
rf.fit(x_train, y_train)
```

```python
rff = rf.predict(x_test) 
accuracy_score(y_test,rff)
```
